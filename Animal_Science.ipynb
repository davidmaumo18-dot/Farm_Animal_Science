```python
animals = [
    {"name": "Bessie", "species": "Cow", "weight_kg": 650, "age_years": 4},
    {"name": "Daisy", "species": "Cow", "weight_kg": 580, "age_years": 2},
    {"name": "Rex", "species": "Dog", "weight_kg": 30, "age_years": 5},
    {"name": "Milo", "species": "Goat", "weight_kg": 45, "age_years": 1},
]

for a in animals:
    print(a["name"], "-", a["species"], "-", a["weight_kg"], "kg")
```

```python
total_weight = sum(a["weight_kg"] for a in animals)
average_weight = total_weight / len(animals)
print("Average weight", round(average_weight, 1), "kg")
```

```python
cows = [a for a in animals if a["species"] == "Cow"]
print("Cows on record:")
for c in cows:
    print("-", c["name"], c["weight_kg"], "kg")
```

```python
def daily_feed_kg(weight_kg, percent=2.5):
    return round(weight_kg * (percent / 100), 2)


for a in animals:
    name = a["name"]
    species = a["species"]
    feed = daily_feed_kg(a["weight_kg"])
    print(name, "(", species, ") needs about", feed,"kg of feed/day")
```

```python
import matplotlib.pyplot as plt


names = [a["name"] for a in animals]
weights = [a["weight_kg"] for a in animals]


plt.bar(names, weights, color="green")
plt.xlabel("Animal")
plt.ylabel("Weight (kg)")
plt.title("Animal Weights")
plt.show()

 
```

```python
growth_log = {
    "Bessie": [600, 620, 640, 650],
    "Daisy": [540, 555, 570, 580],
}

for name, weights in growth_log.items():
    plt.plot(weights, marker="o", label=name)

plt.xlabel("Check-up number")
plt.ylabel("Weight (kg)")
plt.title("Weight Growth Over Time")
plt.legend()
plt.show()
```

```python
target_ranges = {
     "Cow": (550, 700),
     "Dog": (20, 35),
     "Goat": (40, 60),
 }

def health_status(species, weight_kg):
  low, high = target_ranges[species]
  if weight_kg < low:
      return "Underweight"
  elif weight_kg > high:
      return "Overweight"
  else:
      return "Healthy"

for a in animals:
    status = health_status(a["species"], a["weight_kg"])
    print(a["name"], "-", a["species"], "-", a["weight_kg"], "kg ->", status)
    
```
